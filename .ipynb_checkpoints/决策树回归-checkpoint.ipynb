重要参数：
criterion：mse使用均方误差，通过叶子节点均值最小化L2损失
       friedman_mse 费尔德曼均方误差，改进过后的均方误差
       mae绝对平均误差 最小化L1损失
注意，接口score返回的不是均方误差

```python
import numpy as np
from sklearn.tree import DecisionTre
eRegressor
import matplotlib.pyplot as plt
```

```python
rng=np.random.RandomState(1)
X=np.sort(5*rng.rand(80,1),axis=0)
y=np.sin(X).ravel()
y[::5]+=3*(0.5-rng.rand(16))
plt.figure()
plt.scatter(X,y)
```

```python
regr_1=DecisionTreeRegressor(max_depth=2)
regr_2=DecisionTreeRegressor(max_depth=5)
regr_1.fit(X,y)
regr_2.fit(X,y)
```

```python
X_test=np.arange(0,5.0,0.01).reshape(-1,1)
```

```python
y_1=regr_1.predict(X_test)
y_2=regr_2.predict(X_test)
```

```python
plt.figure()
plt.scatter(X,y,s=20,label='data')
plt.plot(X_test,y_1,label='max_depth=2')
plt.plot(X_test,y_2,label='max_depth=5')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Decision Tree Regression')
plt.legend()
plt.show()
```
