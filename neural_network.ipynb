计算神经网络层数时，输入层不算
+ softmax函数：神经网络进行多分类时，默认放在输出层中处理数据的函数
+ 二分类用sigmoid，而多分类不用的原因，虽然可以通过建立多个sigmoid函数来训练，但是神经网络本身就很慢，再建立多个模型会更慢。
   但是，在MLP模型中，sigmoid常作为节点的激活函数
+ z=回归类算法结果，z=wx，k表示k个类别，三个输出，即3个类
这个函数就是一个类别比全部类别

优化该函数，防止数字过大拖延计算,在sklearn中，C是已经设定好的

$\frac{e^{z_k+C'}}{\sum_{i=1}^Ke^{z_i+C'}}$

```python
from sklearn.neural_network import MLPClassifier as MLP
```

参数：
+ activation：每个节点的激活函数
    + relu: f(x)=max(0,x) , tanh: f(x)=tanh(x), logistic:即sigmoid f(x)=1/(1+exp(-x)), identity:f(x)=x
+ solver: 求解器
    + sgd：随机梯度下降
    + lbfgs：牛顿法 ，可用于更小的数据集，收敛更快
    + adam：基于随机梯度的优化方法，在千万级别的数据上表现更好
+ alpah：L2惩罚项，默认0.0001
+ hideen_layer_sizes：中间层体积，默认(100,) 即1个隐藏层，100个神经元
    若(64，64，32）即3个隐藏，分别有64，64，32个神经元
+ max_iter ：最大迭代次数

**容易收敛在局部最优，所以需要多次测试**
